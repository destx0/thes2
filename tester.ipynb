```python
import pandas as pd

df = pd.read_csv('worldcitiespop.csv')

india_cities = df[df['Country'] == 'in']
random_cities = india_cities.sample(n=10000)
unique_cities = random_cities.drop_duplicates(subset=['Latitude', 'Longitude'])

num_cities = unique_cities['City'].nunique()
print(num_cities , unique_cities.columns , unique_cities.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Exponential decay function
def exponential_decay(x, lambda_):
    return np.exp(-lambda_ * x)


# Generating values
x_values = np.linspace(0, 5, 100)
y_values = exponential_decay(x_values, lambda_=1)  # lambda set to 1 for demonstration

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, label="Exponential Decay")
plt.title("Exponential Decay Curve")
plt.xlabel("Distance")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
# Adjusting the exponential decay function for a steeper decay
# Increasing the value of lambda will make the decay steeper

# Generating values with a higher lambda value
lambda_steep = 3  # Increased lambda for steeper decay
y_values_steep = exponential_decay(x_values, lambda_=lambda_steep)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values_steep, label="Exponential Decay ", color="red")
plt.title("Exponential Decay Curve ")
plt.xlabel("Distance")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles

# Generate a circular dataset
X, y = make_circles(n_samples= no_points, factor=0.5, noise=0.05)

# Plotting the dataset
plt.figure(figsize=(6, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="viridis", edgecolor="k")
plt.title("Circular Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
