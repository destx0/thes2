```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import random
```

```python

X, Y = make_blobs(n_samples=100, centers=5, random_state=42, cluster_std=[0.5, 0.7, 0.9, 1.2, 1.5])
```

```python

# Plotting the blobs
# plt.figure(figsize=(8,6))
# plt.figure()
plt.scatter(X[:, 0], X[:, 1] , edgecolors='k')

plt.title('Generated Blobs using make_blobs')
# plt.colorbar()
plt.grid(True)
plt.show()
```

```python
print(type(X))
```

```python
num_edges = 10  # Number of random edges to generate
edges = []

for _ in range(num_edges):
    start_index = random.randint(0, len(X) - 1)
    end_index = random.randint(0, len(X) - 1)
    while end_index == start_index:
        end_index = random.randint(0, len(X) - 1)
    edges.append((start_index, end_index))

# Plotting the blobs and the edges
plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='viridis', s=50)

# Drawing the random edges
for edge in edges:
    x_edge = [X[edge[0], 0], X[edge[1], 0]]
    y_edge = [X[edge[0], 1], X[edge[1], 1]]
    plt.plot(x_edge, y_edge, color='red', linestyle='--', linewidth=0.8)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Blobs with Random Edges')
plt.grid(True)
plt.show()
```
