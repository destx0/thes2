```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
start_value = 10
end_value = 100
n_samples = 10

# Generate exponential distribution
lambda_value = 0.1
samples = np.random.exponential(1/lambda_value, n_samples)

# Scaling and shifting to fit into [start_value, end_value]
scaled_samples = samples * (end_value - start_value) + start_value

# Plot histogram
plt.hist(scaled_samples, bins=30)
plt.title("Exponential Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
from collections import Counter

lpp = np.random.laplace(0, 10, 100000)
lpp = np.abs(lpp).astype(int)
cnt = Counter(list(lpp))
print(cnt)
```

```python
import numpy as np

samples = np.random.standard_cauchy(1000)

x0 = 0  # location parameter
gamma = 2  # scale parameter

scaled_samples = x0 + gamma * samples
lpp = np.abs(scaled_samples).astype(int)
cnt = Counter(list(lpp))
print(cnt)
```
